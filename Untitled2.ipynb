```python
!pip install wfdb heartpy numpy scipy tensorflow scikit-learn matplotlib

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import wfdb
from scipy.signal import butter, filtfilt, savgol_filter
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (Conv1D, MaxPooling1D, LSTM,
                                    Dense, Dropout, Flatten, BatchNormalization,
                                    Bidirectional)
from tensorflow.keras.utils import to_categorical
import heartpy as hp
import time


# Download MIT-BIH records (e.g., records 100, 101, 102)
records = ['100', '101', '102']
wfdb.dl_database('mitdb', dl_dir='mitdb', records=records)

# Load a record
record = wfdb.rdrecord('mitdb/100')
signal = record.p_signal[:, 0]  # Lead 1
fs = record.fs  # Sampling frequency (360 Hz)

# Simulate athlete ECG by adding noise (motion artifacts)
def add_athlete_noise(signal, noise_level=0.05):
    noise = np.random.normal(0, noise_level * np.std(signal), len(signal))
    return signal + noise

athlete_signal = add_athlete_noise(signal)

# Plot original vs athlete ECG
plt.figure(figsize=(15, 5))
plt.subplot(2, 1, 1)
plt.plot(signal[:1000], label='Original ECG')
plt.title('Standard ECG')
plt.subplot(2, 1, 2)
plt.plot(athlete_signal[:1000], label='Athlete ECG (with noise)')
plt.title('Simulated Athlete ECG with Motion Artifacts')
plt.tight_layout()
plt.show()

#######################################

def preprocess_athlete_ecg(signal, fs=360):
    # Bandpass filter (0.5 - 45 Hz)
    nyq = 0.5 * fs
    low, high = 0.5 / nyq, 45.0 / nyq
    b, a = butter(4, [low, high], btype='band')
    filtered = filtfilt(b, a, signal)

    # Savitzky-Golay smoothing (reduces motion noise)
    filtered = savgol_filter(filtered, window_length=21, polyorder=2)

    # Remove baseline wander
    baseline = savgol_filter(filtered, window_length=501, polyorder=3)
    return filtered - baseline

clean_athlete_ecg = preprocess_athlete_ecg(athlete_signal)

# Plot before & after filtering
plt.figure(figsize=(15, 5))
plt.subplot(2, 1, 1)
plt.plot(athlete_signal[:1000], label='Noisy ECG')
plt.title('Raw Athlete ECG (with motion artifacts)')
plt.subplot(2, 1, 2)
plt.plot(clean_athlete_ecg[:1000], label='Filtered ECG')
plt.title('After Noise Removal')
plt.tight_layout()
plt.show()


####################################################################3
# Detect R-peaks using Pan-Tompkins algorithm
def detect_r_peaks(signal, fs=360):
    # Correct way to handle HeartPy 1.2.7+ output
    working_data, measures = hp.process(signal, sample_rate=fs)
    return working_data['peaklist']

r_peaks = detect_r_peaks(clean_athlete_ecg, fs)

# Extract heartbeats (0.6s window around R-peak)
beats = []
window_size = int(0.6 * fs)
half_window = window_size // 2

for r in r_peaks:
    if r - half_window >= 0 and r + half_window < len(clean_athlete_ecg):
        beat = clean_athlete_ecg[r - half_window : r + half_window]
        beats.append(beat)

beats = np.array(beats)

# Plot first 5 heartbeats
plt.figure(figsize=(15, 5))
for i in range(5):
    plt.plot(np.linspace(-0.3, 0.3, window_size), beats[i], label=f'Beat {i+1}')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (mV)')
plt.title('Extracted Heartbeats')
plt.legend()
plt.show()

# Compute HRV features (for athlete performance analysis)
rr_intervals = np.diff(r_peaks) / fs * 1000  # in milliseconds
hrv_features = {
    'mean_rr': np.mean(rr_intervals),
    'sdnn': np.std(rr_intervals),  # Standard deviation of RR intervals
    'rmssd': np.sqrt(np.mean(np.square(np.diff(rr_intervals)))),
    'nn50': np.sum(np.abs(np.diff(rr_intervals)) > 50),
    'pnn50': np.sum(np.abs(np.diff(rr_intervals)) > 50) / len(rr_intervals)
}

print("HRV Features:", hrv_features)

#####################################################
# Load annotations (labels)
ann = wfdb.rdann('mitdb/100', 'atr')

# Map beat types (normal vs ventricular)
symbol_map = {'N': 'normal', 'V': 'ventricular'}
labels = []

# First collect all valid beats and their labels
valid_beats = []
valid_labels = []

for i, r in enumerate(r_peaks):
    # Check if beat extraction was successful
    if i >= len(beats):
        continue  # Skip if no corresponding beat was extracted

    # Get annotation
    nearest_idx = np.argmin(np.abs(ann.sample - r))
    symbol = ann.symbol[nearest_idx]
    label = symbol_map.get(symbol, 'unknown')

    # Only keep normal and ventricular beats
    if label in ['normal', 'ventricular']:
        valid_beats.append(beats[i])
        valid_labels.append(label)

# Convert to numpy arrays
X = np.array(valid_beats)
y = np.array(valid_labels)

# Encode labels (0=normal, 1=ventricular)
le = LabelEncoder()
y_encoded = le.fit_transform(y)
y_onehot = to_categorical(y_encoded)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y_onehot, test_size=0.2, random_state=42)

# Reshape for CNN-LSTM (samples, timesteps, channels)
X_train = X_train[..., np.newaxis]
X_test = X_test[..., np.newaxis]

# Build hybrid CNN-LSTM model
model = Sequential([
    Conv1D(64, 11, activation='relu', input_shape=(X_train.shape[1], 1)),
    BatchNormalization(),
    MaxPooling1D(2),
    Conv1D(128, 7, activation='relu'),
    BatchNormalization(),
    MaxPooling1D(2),
    Bidirectional(LSTM(64, return_sequences=True)),
    Dropout(0.5),
    Bidirectional(LSTM(32)),
    Dense(64, activation='relu'),
    Dense(2, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# Train the model
history = model.fit(X_train, y_train, epochs=15, batch_size=32, validation_data=(X_test, y_test))
# After model training and prediction
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_test, axis=1)

# Check if we have both classes present
unique_classes = np.unique(y_true_classes)
if len(unique_classes) == 1:
    print(f"Warning: Only one class present in test data ({'Normal' if unique_classes[0] == 0 else 'Ventricular'})")
    # Show basic accuracy if only one class
    accuracy = np.mean(y_pred_classes == y_true_classes)
    print(f"Accuracy: {accuracy:.2f}")
else:
    # Full classification report if both classes present
    print(classification_report(
        y_true_classes,
        y_pred_classes,
        target_names=['Normal', 'Ventricular'],
        zero_division=0
    ))

# Plot training history (unchanged)
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.tight_layout()
plt.show()

#########################################################################33
# Simulate real-time ECG stream (5-second window)
def simulate_real_time_ecg(signal, window_size=5*fs, step_size=100):
    for i in range(0, len(signal) - window_size, step_size):
        window = signal[i:i + window_size]
        filtered = preprocess_athlete_ecg(window)
        r_peaks = detect_r_peaks(filtered)

        if len(r_peaks) >= 2:  # At least 2 beats for HRV
            rr_intervals = np.diff(r_peaks) / fs * 1000
            hrv = np.std(rr_intervals)  # SDNN

            # Predict arrhythmia for each beat
            beats = []
            for r in r_peaks:
                if r - half_window >= 0 and r + half_window < len(filtered):
                    beat = filtered[r - half_window : r + half_window]
                    beats.append(beat)

            if beats:  # Changed from If to if
                beats = np.array(beats)[..., np.newaxis]
                preds = model.predict(beats)
                abnormal = np.any(preds[:, 1] > 0.9)  # Ventricular threshold

                if abnormal:  # Changed from If to if
                    print(f"🚨 ALERT: Abnormal rhythm detected at {i/fs:.1f}s (HRV SDNN: {hrv:.1f} ms)")
                else:
                    print(f"✅ Normal at {i/fs:.1f}s (HRV SDNN: {hrv:.1f} ms)")

# Run simulation
simulate_real_time_ecg(athlete_signal)  # Fixed missing closing parenthesis
```
