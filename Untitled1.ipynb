```python
# Install wfdb
!pip install wfdb

import wfdb
import matplotlib.pyplot as plt

# PhysioNet record name
record_name = '100'

# Download record 100 from MIT-BIH
wfdb.dl_database('mitdb', dl_dir='mitdb', records=[record_name])

# Read the downloaded record
record = wfdb.rdrecord(f'mitdb/{record_name}')

# Get the first signal (lead 1)
signal = record.p_signal[:,0]
fs = record.fs

print(f"Sampling rate: {fs} Hz")
print(f"Signal length: {len(signal)} samples")

# Plot first 10 seconds
import numpy as np
plt.figure(figsize=(15,4))
plt.plot(np.arange(0,10*fs)/fs, signal[0:10*fs])
plt.xlabel('Time (s)')
plt.ylabel('mV')
plt.title(f'ECG signal from record {record_name}')
plt.show()

####################################################
from scipy.signal import butter, filtfilt

def bandpass_filter(signal, lowcut=0.5, highcut=45.0, fs=360, order=4):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    filtered = filtfilt(b, a, signal)
    return filtered

filtered_signal = bandpass_filter(signal, fs=fs)

# plot filtered
plt.figure(figsize=(15,4))
plt.plot(np.arange(0,10*fs)/fs, filtered_signal[0:10*fs])
plt.xlabel('Time (s)')
plt.ylabel('mV')
plt.title(f'Filtered ECG signal from record {record_name}')
plt.show()

###########################################################

from wfdb import processing

# Initialize the QRS detector
xqrs = processing.XQRS(sig=filtered_signal, fs=fs)

# Run detection
xqrs.detect()

# Plot the detected R-peaks
plt.figure(figsize=(15,4))
plt.plot(np.arange(0,10*fs)/fs, filtered_signal[0:10*fs])
rpeaks_in_window = xqrs.qrs_inds[xqrs.qrs_inds < 10*fs]
plt.plot(rpeaks_in_window/fs, filtered_signal[rpeaks_in_window], 'ro')
plt.xlabel('Time (s)')
plt.ylabel('mV')
plt.title(f'QRS detections on record {record_name}')
plt.show()


#############################################################

# window length in samples
window_size = int(0.6 * fs)   # 0.6 seconds
half_window = window_size // 2

# store heartbeats here
beats = []

# loop through R-peak positions
for r in xqrs.qrs_inds:
    start = r - half_window
    end = r + half_window
    # avoid out-of-bounds
    if start >= 0 and end <= len(filtered_signal):
        beat = filtered_signal[start:end]
        beats.append(beat)

beats = np.array(beats)

print(f"Extracted {len(beats)} heartbeats, each of shape {beats.shape[1]} samples")

# plot the first heartbeat
plt.figure(figsize=(10,4))
plt.plot(np.linspace(-0.3,0.3,window_size), beats[0])
plt.xlabel("Time (s) around R-peak")
plt.ylabel("mV")
plt.title("First extracted heartbeat")
plt.show()

#################################################

# Load annotations for record 100
ann = wfdb.rdann(f'mitdb/{record_name}', 'atr')

# Print first few annotations
print("First 10 annotations:")
for i in range(10):
    print(f"Sample: {ann.sample[i]}, Symbol: {ann.symbol[i]}")

# Map annotation symbols to classes (AAMI standard)
# N = normal
# V = ventricular ectopic
# S = supraventricular ectopic
# F = fusion
# Q = unknown
symbol_map = {
    'N': 'normal',
    'L': 'normal', # left bundle branch
    'R': 'normal', # right bundle branch
    'A': 'supraventricular',
    'V': 'ventricular',
    'F': 'fusion',
    'E': 'unknown',
    '/': 'paced',
    'f': 'fusion',
    'Q': 'unknown'
}

# Label each heartbeat
labels = []

for r in xqrs.qrs_inds:
    # find nearest annotation
    idx = np.argmin(np.abs(ann.sample - r))
    ann_symbol = ann.symbol[idx]
    class_label = symbol_map.get(ann_symbol, 'unknown')
    labels.append(class_label)

print(f"Labeled {len(labels)} beats")

####################################################
# align sizes of beats and labels
min_len = min(len(beats), len(labels))
beats = beats[:min_len]
labels = labels[:min_len]



from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical

# Convert labels to numeric
le = LabelEncoder()
y_numeric = le.fit_transform(labels)

# only keep normal vs ventricular for a simple binary classification
mask = (np.array(y_numeric) == le.transform(['normal'])[0]) | (np.array(y_numeric) == le.transform(['ventricular'])[0])
X = beats[mask]
y = y_numeric[mask]

# one-hot encoding
y_cat = to_categorical(y)

# reshape for CNN [samples, timesteps, channels]
X = X[..., np.newaxis]

# train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y_cat, test_size=0.2, random_state=42)

print(f"Train samples: {X_train.shape}, Test samples: {X_test.shape}")

###############################################################

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout

model = Sequential([
    Conv1D(16, kernel_size=5, activation='relu', input_shape=(X_train.shape[1], 1)),
    MaxPooling1D(pool_size=2),
    Conv1D(32, kernel_size=3, activation='relu'),
    MaxPooling1D(pool_size=2),
    Flatten(),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(y_cat.shape[1], activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# Train
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))

##################################################################

import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'], label='train acc')
plt.plot(history.history['val_accuracy'], label='val acc')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.title('CNN Training Accuracy')
plt.show()


```
